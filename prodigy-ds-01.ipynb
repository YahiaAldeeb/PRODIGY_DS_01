```python
gender_columns_rural = ['MALE (RURAL)', 'FEMALE (RURAL)']
```

```python
gender_columns_urban = ['MALE (URBAN)', 'FEMALE (URBAN)']
```

```python
gender_distribution_rural = df[gender_columns_rural].sum()
```

```python
gender_distribution_urban = df[gender_columns_urban].sum()
```

```python
plt.figure(figsize=(10, 8))
gender_distribution_rural.plot(kind='bar', color=['blue', 'pink'])
plt.title('Distribution of Population by Gender in Rural Areas')
plt.xlabel('Gender')
plt.ylabel('Population Count')
plt.xticks(rotation=0)
```

```python
plt.figure(figsize=(10, 8))
gender_distribution_urban.plot(kind='bar', color=['blue', 'pink'])
plt.title('Distribution of Population by Gender in Urban Areas')
plt.xlabel('Gender')
plt.ylabel('Population Count')
plt.xticks(rotation=0)
```
